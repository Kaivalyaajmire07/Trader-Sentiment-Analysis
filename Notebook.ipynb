## 📊 Summary

This analysis examines how market sentiment (Fear vs Greed) affects trader performance and behavior...

### 🔹 Methodology  
- Data cleaning and preprocessing  
- Timestamp alignment  
- Feature engineering  

### 🔹 Insights  
- Higher profitability during Fear  
- Increased activity during Fear  
- Larger trades in Greed but lower efficiency  

### 🔹 Strategy Recommendations  
- Increase participation during Fear  
- Reduce position size during Greed
